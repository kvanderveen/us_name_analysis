## What's in a name?  Since 1936 US residents have registered with the Social Security Administration (SSA) and naming data has been collected as a result.  Such naming data is provided online by the SSA and is available for [download](https://www.ssa.gov/oact/babynames/limits.html) at no cost.  In this Jupyter Notebook we explore this name data from which many interesting patterns can be identified.

### First, we need to import the requisite python libraries used for data exploration and analysis.

```python
import glob
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use("fivethirtyeight")
%matplotlib inline
```

### The US name data is provided by the Social Security Administration and is divided into comma-delimited text files, with each individual file representing a separate year (1880-2017).   An example of the structure of each file is shown below.

```python
!head -5 names/yob1880.txt
```

### Using python's glob library and a pattern common to each name file ("names/*.txt"), we can create a list of all files containing name data.

```python
file_names = glob.glob("names/*.txt")
file_names[:5]
```

### Using the list of filenames, a list of pandas dataframes (with each dataframe created from the path of each file name) is created.  Note the creation of a year column using python's re (regular expression) library.  The fact that this takes around 1.2 seconds to run is remarkable.

```python
dataframes = []

for file_name in file_names:
    df = pd.read_csv(file_name, header=None, names=["name", "sex", "count"])
    year = int(re.findall(r"\d{4}", file_name)[0])
    df["year"] = year
    dataframes.append(df)
```

### A single dataframe is then created using the pandas concat function.

```python
df = pd.concat(dataframes)
df.head()
```

### An examination of the total number of names registered with the SSA for each sex by year (1880-2017).  Note the higher number of males after 1936.  It's important to keep in mind that the SSA came into existence in 1935 with the passage of the Social Security Act.  The first registrations did not occur until November 1936.  Therefore, data prior to this time is incomplete.  A variety of explanations could account for the relative higher number of female names registered with the SSA prior to 1937 (which is at odds with the [observation](https://www.npr.org/sections/health-shots/2015/03/30/396384911/why-are-more-baby-boys-born-than-girls) that more male babies are born than females).

```python
number_of_total_names_per_year = df.pivot_table(
    index="year", columns="sex", values="count", aggfunc=np.sum
)
number_of_total_names_per_year.head()
```

```python
ax = number_of_total_names_per_year.plot(
    figsize=(20, 10), fontsize=20, color=["red", "blue"], alpha=0.5, rot=0
)
title = ax.set_title(
    "\nNumber of Baby Names Registered with the SSA 1880-2017\n", fontdict=dict(size=35)
)
xlabel = ax.set_xlabel("\nYear\n", fontdict=dict(size=30))
ylabel = ax.set_ylabel("\nCount\n", fontdict=dict(size=30))
y_tick_format = ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter("{x:,.0f}"))
legend = ax.legend(["Female", "Male"], loc="upper left", fontsize=25)
plt.savefig("figures/annual_count_total_baby_names.png", bbox_inches="tight")
```

### Let's explore the mean character length of baby names registered with the SSA by sex and by year.  Note how the mean name length is longer for females than males and that the name length had been increasing until around 1990-1995.

```python
def mean_string_length_of_series(series):
    lengths = series.str.len()
    return lengths.mean()
```

```python
mean_name_character_lengths = df.pivot_table(
    index="year", columns="sex", values="name", aggfunc=mean_string_length_of_series
)
mean_name_character_lengths.head()
```

```python
ax = mean_name_character_lengths.plot(
    figsize=(20, 10), fontsize=20, color=["red", "blue"], alpha=0.5, rot=0
)
title = ax.set_title(
    "\nMean Character Length of Baby Names Registered with the SSA\n", fontdict=dict(size=35)
)
xlabel = ax.set_xlabel("\nYear\n", fontdict=dict(size=30))
ylabel = ax.set_ylabel("\nCharacter Length\n", fontdict=dict(size=30))
legend = ax.legend(["Female", "Male"], loc="upper left", fontsize=25)
plt.savefig("figures/annual_character_length_baby_names.png", bbox_inches="tight")
```

### Let's examine the most common US baby names registered with the SSA in a given time period.  The 2 functions below allow us to do so.

```python
def generate_top_10_names(start, end, sex):
    top_10 = (
        df[(df.year.isin(range(start, end + 1))) & (df.sex == sex)]
        .groupby("name")["count"]
        .sum()
        .sort_values(ascending=False)
        .iloc[:10]
    )
    return top_10
```

```python
def plot_top_10_names(start, end, sex):
    assert (
        type(start) is int
        and type(end) is int
        and start in range(1880, 2017)
        and end in range(1880, 2018)
        and start <= end
        and sex in ["M", "F"]
    )
    color = {"F": "red", "M": "blue"}.get(sex)
    sex_title = {"F": "Female", "M": "Male"}.get(sex)
    data = generate_top_10_names(start, end, sex)
    ax = data.plot.bar(figsize=(20, 10), fontsize=20, color=color, alpha=0.5, rot=45)
    title = ax.set_title(
        f"\nTop 10 {sex_title} Names Registered with the SSA in {start}-{end}\n",
        fontdict=dict(size=35),
    )
    xlabel = ax.set_xlabel("\nName\n", fontdict=dict(size=30))
    ylabel = ax.set_ylabel("\nCount\n", fontdict=dict(size=30))
    y_tick_format = ax.yaxis.set_major_formatter(
        mpl.ticker.StrMethodFormatter("{x:,.0f}")
    )
    legend = ax.legend("")
```

### A few examples of the most common names in a given time period are seen below.

### Let's explore the gender balance of a name.   Let's limit our search to names with a minimum total count of 1000 or more.

```python
name_counts_by_sex = (
    df.pivot_table(index="name", columns="sex", values="count", aggfunc=np.sum)
    .fillna(0)
    .astype(int)
)
name_counts_by_sex = name_counts_by_sex[name_counts_by_sex.sum(axis="columns") >= 1000]
name_counts_by_sex.head()
```

### To identify gender balanced names we calculate the absolute values of the differences between female and male counts for each name divided by the total count of each name.  Values of this calculation close to zero suggest a name is gender balanced.  The name Ocean represents the most gender balanced name with a count of at least 1000. 

```python
idx = (
    (
        (name_counts_by_sex.F - name_counts_by_sex.M).abs()
        / name_counts_by_sex.sum(axis="columns")
    ).sort_values()
).index
```

```python
top_10_balanced_names = data = name_counts_by_sex.loc[idx[:10], :]
top_10_balanced_names.head()
```

```python
ax = top_10_balanced_names.plot.bar(
    figsize=(20, 10), fontsize=20, color=["red", "blue"], alpha=0.5, rot=45
)
title = ax.set_title(
    '\nCount of the Most Gender Balanced Names\n', fontdict=dict(size=35)
)
xlabel = ax.set_xlabel("\nName\n", fontdict=dict(size=30))
ylabel = ax.set_ylabel("\nCount\n", fontdict=dict(size=30))
y_tick_format = ax.yaxis.set_major_formatter(
        mpl.ticker.StrMethodFormatter("{x:,.0f}")
    )
legend = ax.legend(["Female", "Male"], loc="upper left", fontsize=25)
plt.savefig("figures/gender_neutral_names.png", bbox_inches="tight")
```

### Another way to evaluate the diversity of names is to examine the percentage of all names the top 100 names represent.  Note that until around 1950, the top 100 names represented a much higher percentage of all names than in recent years.  Consistent with an earlier observation of the greater variety of female names than male names, the top 100 female names consistently represent a smaller percentage of total female names than analogously measured for males.

```python
def percent_is_top_100(series):
    series = series.sort_values(ascending=False)
    top_100 = series[:100].sum()
    return round(top_100 / series.sum() * 100, 1)
```

```python
percent_top_100 = df.pivot_table(
    index="year", columns="sex", values="count", aggfunc=percent_is_top_100
).sort_index(ascending=True)
percent_top_100.head()
```

```python
ax = percent_top_100.plot(
    figsize=(20, 10), fontsize=20, color=["red", "blue"], alpha=0.5, rot=0
)
title = ax.set_title(
    "\nPercent of All Names Represented in the Top 100 Names\n", fontdict=dict(size=35)
)
xlabel = ax.set_xlabel("\nYear\n", fontdict=dict(size=30))
ylabel = ax.set_ylabel("\nPercent\n", fontdict=dict(size=30))
y_tick_format = ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter("{x:.0f}%"))
legend = ax.legend(["Female", "Male"], loc="upper right", fontsize=25)
plt.savefig("figures/percent_of_top_100_names.png", bbox_inches="tight")
```
